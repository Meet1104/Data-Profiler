## 1. What is Data Analysis?

### Definition
Data Analysis is the process of inspecting, cleaning, transforming, and modeling data with the goal of discovering useful information, drawing conclusions, and supporting decision-making. It is a crucial step before building any machine learning or AI system, as it helps us understand the story hidden inside data.

### Purpose
The main purpose of data analysis is to:

- Understand the structure and quality of the dataset  
- Identify patterns, trends, and relationships within the data  
- Detect anomalies or outliers that may impact results  
- Support business decisions with data-driven insights  

### Objective

After studying this section, learners should be able to:

- Explain what data analysis is and why it is important  
- Differentiate between descriptive, diagnostic, predictive, and prescriptive analysis  
- Apply basic Python tools (like Pandas, NumPy, Matplotlib) to perform Exploratory Data Analysis (EDA)  
- Make preliminary insights from raw data  

### Syntax / Implementation (Python)

In Python, data analysis usually starts with importing required libraries:

- import pandas as pd
- import numpy as np    
- import matplotlib.pyplot as plt  
- import seaborn as sns  

### Real-World Example 🌍

Imagine a retail store wants to know which products sell the most during festivals. By analyzing past sales data, they can:

- Identify the top-selling products  
- Plan stock levels to avoid shortages  
- Launch targeted promotions for specific items  

### Technical Example ⚙️

In Machine Learning, before training a model on a dataset, we analyze:

- Distribution of features (e.g., age, income, product price)  
- Missing values and their handling strategy  
- Correlation between independent variables and the target variable  


## 2. Steps to Plan a Data Science Project.

A typical Data Science project follows these steps:

1. **Problem Definition**
   - Clearly define the objective (e.g., predict customer churn)

2. **Data Collection**
   - Gather data from sources like databases, APIs, CSV files

3. **Data Cleaning**
   - Handle missing values, duplicates, and inconsistencies

4. **Exploratory Data Analysis (EDA)**
   - Understand patterns, distributions, and relationships

5. **Feature Engineering**
   - Create new features and select important ones

6. **Model Building**
   - Choose and train machine learning models

7. **Model Evaluation**
   - Evaluate using metrics like accuracy, precision, recall

8. **Deployment**
   - Deploy the model in real-world applications

9. **Monitoring & Maintenance**
   - Track performance and update when needed

## 4. Explain: What are Tensors?

### Definition
A Tensor is a multi-dimensional array or data structure used to represent data in mathematics, physics, and machine learning. It is a generalization of scalars, vectors, and matrices to higher dimensions.

### Types of Tensors

- **Scalar (0D Tensor)**  
  A single value  
  Example: 5  

- **Vector (1D Tensor)**  
  A list of values  
  Example: [1, 2, 3]  

- **Matrix (2D Tensor)**  
  A 2D table of values  
  Example:  
  [[1, 2],  
   [3, 4]]  

- **Higher-Dimensional Tensor (3D and above)**  
  Example: Data used in images, videos, etc.

---

### Explanation with NumPy Examples

In Python, tensors can be represented using NumPy arrays:

```python
import numpy as np

# Scalar (0D Tensor)
scalar = np.array(10)
print("Scalar:", scalar)

# Vector (1D Tensor)
vector = np.array([1, 2, 3])
print("Vector:", vector)

# Matrix (2D Tensor)
matrix = np.array([[1, 2], [3, 4]])
print("Matrix:\n", matrix)

# 3D Tensor
tensor_3d = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print("3D Tensor:\n", tensor_3d)
